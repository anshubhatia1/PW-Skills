### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an Example.

#### Probability Mass Function (PMF):
- **Definition:** The PMF is a function that gives the probability that a discrete random variable is exactly equal to some value. It maps each possible value of the discrete random variable to the probability of that value occurring.
- **Formula:**
  \[
  P(X = x) = f(x)
  \]
  where \( X \) is a discrete random variable, \( x \) is a possible value of \( X \), and \( f(x) \) is the probability of \( x \).

- **Example:** Consider a fair six-sided die. The PMF of the die roll is:
  \[
  P(X = x) = \begin{cases} 
  \frac{1}{6} & \text{if } x \in \{1, 2, 3, 4, 5, 6\} \\
  0 & \text{otherwise}
  \end{cases}
  \]

#### Probability Density Function (PDF):
- **Definition:** The PDF is a function that describes the likelihood of a continuous random variable taking on a particular value. The value of the PDF at any given point can be thought of as the relative likelihood that the value of the random variable is near that point.
- **Formula:**
  \[
  f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2}
  \]
  where \( \mu \) is the mean, \( \sigma \) is the standard deviation, and \( x \) is the point at which the PDF is evaluated.

- **Example:** For a normal distribution with mean 0 and standard deviation 1 (standard normal distribution), the PDF is:
  \[
  f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} x^2}
  \]

### Q2: What is Cumulative Density Function (CDF)? Explain with an Example. Why is CDF used?

#### Cumulative Density Function (CDF):
- **Definition:** The CDF of a random variable \( X \) is a function that gives the probability that \( X \) will take a value less than or equal to \( x \). It provides a cumulative sum of the probabilities up to a given point.
- **Formula:**
  \[
  F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt
  \]
  where \( f(t) \) is the PDF for continuous variables or the PMF for discrete variables.

- **Example:** For a standard normal distribution, the CDF is:
  \[
  F(x) = \int_{-\infty}^{x} \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} t^2} \, dt
  \]
  This integral does not have a closed-form solution and is typically computed using numerical methods or looked up in statistical tables.

- **Use of CDF:**
  - The CDF is used to determine the probability that a random variable takes on a value less than or equal to a specific value.
  - It is useful in statistical analysis for calculating probabilities and quantiles.

### Q3: Examples of Situations Where the Normal Distribution Might Be Used as a Model

#### Examples:
1. **Heights of people:** Human heights tend to follow a normal distribution.
2. **IQ scores:** IQ scores are often modeled as a normal distribution with a mean of 100 and a standard deviation of 15.
3. **Measurement errors:** Errors in measurements due to instrument precision often follow a normal distribution.

#### Parameters and Shape:
- **Mean (μ):** Determines the center of the distribution.
- **Standard Deviation (σ):** Determines the spread or width of the distribution. A larger standard deviation means a wider distribution.
- **Shape:** The normal distribution is symmetric and bell-shaped, centered around the mean.

### Q5: What is Bernoulli Distribution? Example and Difference Between Bernoulli and Binomial Distribution

#### Bernoulli Distribution:
- **Definition:** A distribution of a random variable which has only two possible outcomes, typically 0 (failure) and 1 (success).
- **Parameter:** Probability of success \( p \).
- **Example:** Tossing a coin where 1 represents heads (success) and 0 represents tails (failure).

#### Difference Between Bernoulli and Binomial Distribution:
- **Bernoulli Distribution:**
  - One trial
  - Two possible outcomes
  - PMF: \( P(X = x) = p^x (1-p)^{1-x} \), for \( x \in \{0, 1\} \)

- **Binomial Distribution:**
  - Multiple trials (\( n \))
  - Each trial is a Bernoulli trial
  - PMF: \( P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \), where \( k \) is the number of successes

### Q6: Probability of a Randomly Selected Observation Greater than 60 from a Normal Distribution with Mean 50 and Standard Deviation 10

To find the probability that a randomly selected observation \( X \) is greater than 60, we need to use the Z-score formula and look up the corresponding value in the standard normal distribution table.

1. **Z-score Calculation:**
   \[
   Z = \frac{X - \mu}{\sigma} = \frac{60 - 50}{10} = 1
   \]

2. **Standard Normal Distribution:**
   Using the Z-table, the probability that \( Z \leq 1 \) is approximately 0.8413. Therefore, the probability that \( Z > 1 \) is:
   \[
   P(Z > 1) = 1 - 0.8413 = 0.1587
   \]

   So, the probability that a randomly selected observation is greater than 60 is 0.1587.

### Q7: Explain Uniform Distribution with an Example

#### Uniform Distribution:
- **Definition:** A distribution where all outcomes are equally likely within a certain range. For a continuous uniform distribution, the PDF is constant between the minimum \( a \) and maximum \( b \) values.
- **PDF:**
  \[
  f(x) = \begin{cases} 
  \frac{1}{b - a} & \text{for } a \leq x \leq b \\
  0 & \text{otherwise}
  \end{cases}
  \]

#### Example:
- Rolling a fair die: Each outcome (1, 2, 3, 4, 5, 6) is equally likely, so the distribution of outcomes is uniform.

### Q8: What is the Z-Score? Importance of the Z-Score

#### Z-Score:
- **Definition:** A Z-score measures how many standard deviations an element is from the mean of the distribution.
- **Formula:**
  \[
  Z = \frac{X - \mu}{\sigma}
  \]
  where \( X \) is the value, \( \mu \) is the mean, and \( \sigma \) is the standard deviation.

#### Importance:
- **Standardization:** Allows comparison of data from different distributions.
- **Outliers:** Helps in identifying outliers.
- **Probability:** Used in calculating probabilities and percentiles in the standard normal distribution.

### Q9: What is the Central Limit Theorem? Significance of the Central Limit Theorem

#### Central Limit Theorem (CLT):
- **Definition:** The CLT states that the distribution of the sample mean of a large number of independent, identically distributed variables approaches a normal distribution, regardless of the shape of the original distribution.
- **Formula:** If \( X_1, X_2, \ldots, X_n \) are i.i.d. random variables with mean \( \mu \) and variance \( \sigma^2 \), then the sample mean \( \bar{X} \) is approximately normally distributed for large \( n \):
  \[
  \bar{X} \sim N \left( \mu, \frac{\sigma^2}{n} \right)
  \]

#### Significance:
- **Statistical Inference:** Provides the foundation for many statistical tests and confidence intervals.
- **Simplifies Analysis:** Allows the use of normal distribution to approximate the behavior of sample means.

1. **Independence:** The sampled observations must be independent.
2. **Sample Size:** The sample size should be sufficiently large (usually \( n \geq 30 \) is considered adequate).
3. **Identical Distribution:** The observations must come from identically distributed populations (i.i.d).
